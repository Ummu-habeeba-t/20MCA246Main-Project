## What is Ecommerce?

|       |       |

|<span style= 'background:yellow'>1. Business to Consumer (B2C)</span>| When a business sells a good or service to an individual consumer (e.g. You buy a pair of shoes from an online retailer).|

# 6. Modeling

## 6.1 Modeling with the linear model and tree based model technique with Standard Scaler.

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scale = scaler.transform(X_train)
```

```python
X_test_scale = scaler.transform(X_test)
```

# Fit/Train the Model

```python
# Choosing the best model among the tree based models
MAE_test_scale_1 = []
MAE_train_scale_1 = []
RMSE_test_scale_1 = []
RMSE_train_scale_1 =[]

MLA= [ LinearRegression(),Ridge(), Lasso(),ElasticNetCV()]

for i in MLA:
    model1=i.fit(X_train_scale ,y_train)
    # evaluate the model
    test_pred_scale_1 = model1.predict(X_test_scale)
    train_pred_scale_1 = model1.predict(X_train_scale)
    # evaluate predictions
    MAE_test_scale_1.append(mean_absolute_error(y_test,test_pred_scale_1))
    MAE_train_scale_1.append(mean_absolute_error(y_train, train_pred_scale_1))
    RMSE_test_scale_1.append(np.sqrt(mean_squared_error(y_test,test_pred_scale_1)))
    RMSE_train_scale_1.append(np.sqrt(mean_squared_error(y_train,train_pred_scale_1)))
```

```python
result_df_scale_1= pd.DataFrame(MAE_test_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'], 
                              columns=['MAE_test']) 
result_df_scale_2= pd.DataFrame(MAE_train_scale_1, index=['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['MAE_train'])
result_df_scale_3= pd.DataFrame(RMSE_test_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['RMSE_test'])
result_df_scale_4= pd.DataFrame(RMSE_train_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['RMSE_train'])
```

```python
# Choosing the best model among the tree based models
MAE_test_scale_1 = []
MAE_train_scale_1 = []
RMSE_test_scale_1 = []
RMSE_train_scale_1 =[]

MLA= [ LinearRegression(),Ridge(), Lasso(),ElasticNetCV()]

for i in MLA:
    model1=i.fit(X_train_scale ,y_train)
    # evaluate the model
    test_pred_scale_1 = model1.predict(X_test_scale)
    train_pred_scale_1 = model1.predict(X_train_scale)
    # evaluate predictions
    MAE_test_scale_1.append(mean_absolute_error(y_test,test_pred_scale_1))
    MAE_train_scale_1.append(mean_absolute_error(y_train, train_pred_scale_1))
    RMSE_test_scale_1.append(np.sqrt(mean_squared_error(y_test,test_pred_scale_1)))
    RMSE_train_scale_1.append(np.sqrt(mean_squared_error(y_train,train_pred_scale_1)))
```

```python
result_df_log_out_1= pd.DataFrame(MAE_test_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'], 
                              columns=['MAE_test']) 
result_df_log_out_2= pd.DataFrame(MAE_train_scale_1, index=['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['MAE_train'])
result_df_log_out_3= pd.DataFrame(RMSE_test_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['RMSE_test'])
result_df_log_out_4= pd.DataFrame(RMSE_train_scale_1, index= ['LinearRegression', 'Ridge', 'Lasso', 
                                                  'ElasticNetCV'],
                              columns=['RMSE_train'])
```
